### ⭐ Introduccion

Imaginemos que queremos buscar las raices de $x^2+1$ (Es decir, los x que hacen que la ecuacion valga cero). Me encontraria con la siguiente ecuacion :
$$ x^2 + 1 = 0 $$
$$x^2 = -1$$
Esto no tiene solucion real, **porque todo numero real al cuadrado es positivo**.

Pero... **No es la primera vez que un conjunto numerico no tiene solucion para una operacion**, no? Es decir
En los naturales nos pasaba (por ejemplo) que $3-5= -2$. Pero $-2 \not \in \mathbb{N}$. Esto significa que **Los naturales no son cerrados para la resta**.

Entonces armamos los numeros enteros $\mathbb{Z}$. Pero, volviamos a tener un problemas, como $\frac{3}{2} \not \in \mathbb{Z}$. **.
Los enteros no son cerrados para la division**.

... Y asi se fueron construyendo conjuntos numericos hasta llegar a $\mathbb{R}$.

Pero ahora tenemos el problema de que $x^2 = -1 \not \in \mathbb{R}$. **Los reales no son cerrados para el exponente (^).**

Para solucionarlo, decimos que $\sqrt{-1} = i$, donde $i$ es llamada **unidad imaginaria**.

Si sumamos un numero real con otro imaginario obtenemos un **numero complejo** $\mathbb{C}$.

$$z = a + bi$$

con $\ a,b \in \mathbb{R}$ y ademas $i \not \in \mathbb{R}$

Como $a$ y $b$ $\in \mathbb{R}$, claramente ambos pueden valer 0 y estaria todo bien (De hecho $0 \in \mathbb{C} $).

Entonces estamos en condiciones de decir que $z \in \mathbb{C}$;

y volviendo al problema anterior ($x^2 = -1$) ya tenemos solucion!
$x = i$
$$ i^2 = -1$$


⚠️ Obs: vamos a decir que $i^2 = -1$ en lugar de $i = \sqrt{-1}$. Ya que **la raiz cuadrada solo existe para los numeros reales**:
$$\sqrt{x} \ : \  \mathbb{R} \rightarrow \mathbb{R}$$

### Teorema fundamental del algebra
Mejor aun! No solo el polinomio anterior va a tener solucion, si no que todos los polinomios van a tener raices siempre y cuando estemos trabajando en $\mathbb{C}$. Eso es lo que dice el **teorema fundamental del algebra**.

"Todo polinomio de grado mayor que cero tiene una raíz.​ El dominio de la variable es el conjunto de los números complejos, que es una extensión de los números reales."

De hecho, la cantidad de raices de un polinomio va a ser igual al grado de ese polinomio. Esto lo veremos mas adelante.

En la siguiente tabla podemos ver como los complejos son cerrados para todas las operaciones. Mientras que los demas conjuntos numericos no.
$$

### ⭐ El conjunto de los complejos

Entonces introdujimos un nuevo conjunto llamado $\mathbb{C}$, el conjunto de los números complejos. El cual cumple que $\mathbb{R} \subset \mathbb{C}$. Vamos a ver las distintas maneras de expresar un numero complejo.

### ⭐ Forma binomial

Dado $z \in \mathbb{C}$ la forma $z = a+b.i$ con $a,b \in \mathbb{R}$ se llama forma binomial de $z$. $\mathbb{C} = \{ z=a+b.i; \ a,b \in \mathbb{R} \}$


#### El plano complejo
Para graficar a los numeros complejos en el plano, no podemos hacerlo mas con el plano real al que nos acostumbramos ($\mathbb{R}^2$). En su lugar, vamos a utilizar el **plano complejo**. Este consta de una **recta imaginaria** que va en el eje $y$ y una **recta real** que va en el eje $x$. Esto guarda similitudes con graficar vectores. La parte real la ubicamos en el eje real y la parte imaginaria en el eje imaginario.

Tambien existe el **conjugado** de un numero complejo, se nota con una barrita arriba: $\bar{z}$.
El conjugado cambia el simbolo del medio de la ecuacion (es decir, el del imaginario).
Ejemplo: 
$z = a + bi$

$\bar{z}= a - bi$

### ⭐ Forma trigonometrica
Es otra manera de notar un numero complejo, podemos deducir esta formula mediante las identidades trigonometricas de la forma binomial.

$a = cos(\theta) . r$

$b = sen(\theta).r$

Donde $r$ = $|z|$ y $\theta$ es el argumento.

⚠️ Para que $\theta$ sea el argumento tiene que cumplir la condicion 
$0 \leq \theta \leq 2\pi$

$$z = r.cos(\theta) + i.sen(\theta).r$$


### ⭐ Notacion exponencial

Esta es otra manera de representar un numero complejo de manera trigonometrica. Proviene de un resultado en el **analisis complejo**.


$$r.e^{\theta i} = r.(\cos(\theta) + i \sin(\theta)) \ \forall \theta \in \mathbb{R}$$
Siendo $r$ el modulo del numero complejo y $\theta$ el argumento (en radianes). A medida que vamos variando el argumento, dariamos una vuelta completa.
Ejemplo de un numero complejo variando $\theta$ con $r = 1$.

### Propiedades modulo y argumento
$r.e^{\theta i}  \ s.e^{\varphi i} = rse^{(\theta+\varphi)i}$

$arg(z.w) = arg(z)+arg(w) - 2k \pi, k \in \mathbb{Z} $

$\arg(z^n) = n \arg(z) -2k\pi < 2\pi, k \in \mathbb{Z} $

$|z .w| = |z|.|w|$

$|z^n| = |z|^n$

## El orden de los numeros complejos
Una particularidad de los numeros complejos es que se pierde el orden que teniamos en los numeros reales.

Se pierde que en $\mathbb{C}$ no se puede establecer ningun orden $\geq$

si $i ≥ 0$ entonces $i^2 = −1 ≥ 0$ implica $0 = −1+1 ≥ 0+1 = 1$ 

pero por otro lado, $1 = (−1)^2 ≥ 0^2 = 0$ . Es decir $0 ≥ 1$ y $1 ≥ 0$ . 

Por la antisimetria, eso tendriıa que implicar $0 = 1$ $\red{\text{absurdo}}$.

Cuando trabajabamos en $\mathbb{R}$, si yo multiplicaba 2 numeros $a$ y $b$ mayores a 0 (positivos), obtenia otro numero positivo. Esto no pasa en los complejos, ej: 
$i . i = -1$
