# Divisibilidad

$a | b$ Significa que a divide a b

Decir que $n$ **divide** a $m$ es equivalente a   decir que   $m$ es un **múltiplo** de $n$

$r_ab = 0$  Significa que el resto de dividir $a$ por $b$ es 0

Y podemos unir las 2 expresiones anteriores: 

$$a|b \leftrightarrow r_ab=0$$

#### La divisibilidad es una relacion de orden sobre N

- **Reflexividad:** $m \ |\ m$
- **Transitividad:** $m \ | \  n$ y $n \ | \ p \rightarrow m \ | \ p$
- **Anti simetría:** $m  \ | \ n$  y $n \ | \ m \rightarrow m=n$

Siendo que la division es transitiva, podemos observar el siguiente fenomeno:

$$ 1 \rightarrow 2 \rightarrow 4 \rightarrow 8 \rightarrow 16 $$ 

Lo que expresa este grafico es que 1 divide a 2, luego 2 a 4, luego 4 a 8 y 8 a 16.
Esto genera una especie de "cadena de divisiones" en la que cada numero divide a todo lo que esta delante de el. El 2 no solo divide al 4, si no tambien al 8 y al 16.
El 4 no solo divide al 8 si no tambien al 16.

Podemos observar unos esquemas muy interesantes que generan la divisibilidad:

$$\begin{matrix}
9  & \longrightarrow & 18 & \longrightarrow & 36 & \longrightarrow & 72 \\
\uparrow & &\uparrow  & &\uparrow  & &\uparrow  \\
3 & \longrightarrow  & 6 & \longrightarrow & 12 & \longrightarrow & 24 \\
\uparrow  & &\uparrow  & &\uparrow  & &\uparrow  \\
1 & \longrightarrow & 2  & \longrightarrow & 4 & \longrightarrow & 8
\end{matrix}$$
<center>Esto se llama <em>Diagrama de Hasse</em></center>


### Definiendo la division formalmente
Sean $a,d \in \mathbb{Z}$ con $d \not =0$. Se dice que $a$ es divisible por $d$ o que a es múltiplo de $d$:

$d \ | \ a \Leftrightarrow \exists \ k \in \mathbb{Z} \ : a=k.d$

#### La congruencia
$a \equiv b \mod d \ $ Es lo mismo que decir $d \ | \ a-b$
