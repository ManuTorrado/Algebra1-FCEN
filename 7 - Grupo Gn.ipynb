# Raices enesimas de z
Planteo el siguiente problema, tengo: 

$z= -\sqrt{2}-\sqrt{2}i$ 

y quiero hallar los
$w^6= z$ 

**Como lo hago?**

Como se que $w \in \mathbb{C}$ entonces $w$ tiene que poder representarse como
$w = re^{\theta i}$

y $w^6 = r^6 e^{6\theta i}$ .

Pero $w^6 = z$, entonces  $ r^6 e^{6\theta i} = -\sqrt{2}-\sqrt{2}i$

$|z| = \sqrt{  (-\sqrt{2})^2+(-\sqrt{2})^2 } = \sqrt{4} = 2 $



$\cos{\varphi} = \frac{-\sqrt{2}}{|z|}$,
$\sin{\varphi} = \frac{-\sqrt{2}}{|z|}$

⚠️ $\varphi = \frac{5}{4}\pi$   **Por que?**

⚠️ Hay que tener cuidado a la hora de ubicar $\varphi$, si bien
$\cos(\frac{-\sqrt{2}}{2}) = \frac{3}{4}\pi$

El $\frac{3}{4}\pi$ esta en sentido **horario** al circulo, necesitamos el anglulo tomado en sentido **anti horario**




Y la forma trigonometrica de $z$ es 

$z = |z|(\cos{\varphi} + i\sin{\varphi})$

$w^6= |z|e^{\varphi}i = r^6 e^{6\theta i}$


$r^6 = |z| ,   6\theta = \varphi + 2k\pi$


$r = \sqrt[6]{2} $

$\theta = \frac{\varphi}{6}+\frac{2k\pi}{6} = \frac{\varphi}{6} + \frac{k\pi}{3} = \frac{5}{4}\pi \frac{1}{6} + \frac{k\pi}{3} = \frac{5}{24}\pi + \frac{k\pi}{3}$

$$w^6 = \sqrt[6]{2} e^{\theta i} $$

# El grupo $G_n$
El grupo $G_n$ son todas las raices de la unidad y se define como:

$$G_{n}=\left\{\omega \in \mathbb{C}: \omega^{n}=1\right\}$$

O tambien como:

$$G_{n}=\left\{\omega_{k}=e^{\frac{2 k \pi}{n} i}, 0 \leq k \leq n-1\right\} $$


**Obs**: $G_n\subseteq \mathbb{C}$


Recordemos que $w^n$ = 1 es un polinomio y como tal tendra $n$ soluciones.

Si buscamos $G_1 = \set{w \in \mathbb{C} / w^1 = 1}$ Esto solo tiene una solucion, 1.

Si buscamos $G_2 = \set{w \in \mathbb{C} / w^2 = 1}$ Esto tiene 2 soluciones, $\set{1,-1}$.

Con $G_3$ tengo que tener 3 soluciones.

El conjunto $G_n$ tiene $n$ elementos distintos en $\mathbb C$ que forman un $n$-ágono regular en la circunferencia unidad del plano complejo, empezando desde el 1.

### Propiedades 

- $\omega \in G_{n} \implies |\omega|=1$
- $\omega \in G_n \Rightarrow \omega^m \in G_n, \forall  m \in \mathbb{Z}$
- $m\equiv m' (n) \Rightarrow \omega^m = w^{m'}$
- $\omega^m = \omega^{r_n(m)}$


## Raices primitivas de la unidad
Sea $n \in \mathbb{N}$ se dice que $\omega \in \mathbb{C}$ es una raiz $n$-esima **primitiva** de la unidad si

$$G_n = \set{1, \omega,..., \omega^{n-1}}  = \set{\omega^k, 0 \leq k \leq n - 1} $$

A veces se suele notar a las raices primitivas como $G_n^*$, y se define con una particularidad:

$$G_n^* = \set{w^k : 1 \leq k \leq n, (k \perp n)} $$
$k$ tiene que ser coprimo con $n$.

#### $\varphi$ de Euler
$ |G^*_n| = \varphi (n) \ =\#\set{\text{Coprimos con }n, \leq n} $


$\varphi(mn)= \varphi(m) \cdot\varphi(n)$ Si $m \perp n$
